# Experiment A02 pKa of Phenols

```python
import pandas as pd
import numpy as np
import os
```

## pKa of 4-iodophenol

```python
#list coprehension to read all files in the directory
files = [file for file in os.listdir('./4-iodophenol')]
for file in files:
    print(file)
```

```python
#empty data frame where all files will be merged
all_data = pd.DataFrame()

#concatenate into single data frame, axis = 1 so that the files are added into new columns instead of rows
for file in files:
    df = pd.read_csv('./4-iodophenol/'+file)
    all_data = pd.concat([all_data, df], axis = 1)
    
#create a new csv file with all data
all_data.to_csv('all_data.csv', index = False)
```

```python
all_data = pd.read_csv('all_data.csv')
all_data.head(10)
```

```python
all_data.columns = ['w_buffer', 'A_buffer', 'w_HCl', 'A_HCl', 'w_NaOH', 'A_NaOH']
all_data.head()
```

```python
import plotly.express as px
fig = px.scatter(x=all_data.w_NaOH, y=all_data.A_NaOH )
fig.add_scatter(x=all_data.w_buffer, y=all_data.A_buffer)
fig.add_scatter(x = all_data.w_HCl, y=all_data.A_HCl)
fig.update_layout(
    title="UV-Vis spectra of 4-iodophenol",
    xaxis_title="wavelength /nm ",
    yaxis_title="Absorption")
fig.show()
```

```python
#for range betwwen 250 and 350nm
range_all_data=all_data.loc[(all_data['w_NaOH'] >= 250) & (all_data['w_NaOH'] <= 350)]
```

```python
import plotly.express as px
fig = px.scatter(x=range_all_data.w_NaOH, y=range_all_data.A_NaOH )
fig.add_scatter(x=range_all_data.w_buffer, y=range_all_data.A_buffer, showlegend = True)
fig.add_scatter(x = range_all_data.w_HCl, y=range_all_data.A_HCl)
fig.update_layout(
    title="UV-Vis spectra of 4-iodophenol between 250 and 350nm",
    xaxis_title="wavelength /nm ",
    yaxis_title="Absorption")
fig.show()
```

```python
d_A_HCl_NaOH = range_all_data.A_HCl -range_all_data.A_NaOH
d_A_HCl_NaOH.max()
print(d_A_HCl_NaOH[d_A_HCl_NaOH==d_A_HCl_NaOH.max()].index.values)
range_all_data.w_HCl[387] #wavelength for which A_HCl-A_NaOH is the greatest
```

```python
from math import log
#calculating x
x = (range_all_data.A_buffer[465]-range_all_data.A_HCl[465])/(range_all_data.A_NaOH[465]-range_all_data.A_HCl[465] )

print(x)
pka_m_i = 8.9 + log((1-x)/x)
print(pka_m_i)
```
